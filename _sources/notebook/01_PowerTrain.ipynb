## TD - Modélisation de la transmission de puissance d’un utilitaire électrique

Le Goupil G4 (**Figure 1**) est un petit utilitaire électrique urbain,
utilisé par les villes et les entreprises pour déplacer de petites
charges sur de courtes distances (livraison, entretien de parcs,
services techniques).

L’objectif de ce problème est de simuler la transmission de puissance de
la motorisation électrique du Goupil. Le **Tableau 1** donne les
principales caractéristiques des différents composants mécaniques du
Goupil G4.

**Modélisation du Goupil**

1.  Proposez un diagramme 1D et donnez les équations correspondantes
    permettant de modéliser la dynamique du Goupil en translation en
    prenant notamment en compte les forces suivantes:

-   L’inertie en translation du Goupil et du conducteur;

<!-- -->

-   L’effet de la gravité dans une pente;

-   Les efforts aérodynamiques
    $F_{aero} = \frac{1}{2}.\rho_{air}C_{x}S.v^{2} \approx 0,5061.v^{2}$
    en N où *v* correspond à la vitesse du Goupil en m/s ;

-   Les efforts de résistance au roulement proportionnels à l’effort
    normal sur les roues et au coefficient de résistance de roulement
    *C<sub>rr</sub>* des pneus. Sur route, avec un gonflage adéquat, on
    considèrera *C<sub>rr</sub>* = 100.10<sup>-4</sup>.

1.  Implémentez ce modèle dans la brique vide du modèle *Goupil*.
    L’entrée de type signal correspond à la pente (où 0.07 correspond à
    une pente de 7%). Le port mécanique permettra de modéliser les
    efforts équivalents de la motorisation.

<u>Remarque 1</u> : *GoupilMass, Crr, Alpha (0,3), g* sont des
coefficients declarés dans le modèle *Goupil*.

<u>Remarque 2</u> : Pour votre modèle vous pouvez utiliser des modèles
de la librairie standard (MSL) comme

-   dans *Modelica/Mechanics/Tranlational/* les sources d’effort
    (*Sources/Force*) et composants de type inertie ou frottement
    (*Components/Mass, Components/SupportFriction*) ;

-   dans *Modelica/Blocks/Math* : *Sum, Product, Gain, Atan, Sin, …*

1.  Calculez analytiquement l’effort à fournir pour obtenir en régime
    permanent une vitesse *v* = 30 km/h sur une pente de 7%.

2.  Validez par simulation ce résultat en utilisant le modèle
    *TestGoupil*.

**Modélisation de la transmission de puissance**

L’ensemble des deux réducteurs (différentiel et à pignons droits) peut
se représenter par un réducteur équivalent présentant les
caractéristiques suivantes :

-   Inertie sur l’axe rapide : 0,57 kg.cm<sup>2</sup> ;

-   Rapport de réduction global : 14.87 ;

-   Rendement global : 88% ;

-   Les effets des raideurs ne sont pas pris en compte ici.

1.  Calculez analytiquement pour le point opérationnel précédent
    (vitesse constante de 30 km/h, pente de 7%) :

-   La vitesse angulaire de la roue et la vitesse du moteur électrique
    (sur l’axe rapide du réducteur équivalent) ;

-   Les pertes du réducteur ;

-   La puissance mécanique sur l’axe rapide ;

-   Le couple et la puissance du moteur ;

1.  Paramétrez et simulez le modèle *TestSlopeMotor*. Ce modèle
    représente un profil de mission que le Goupil doit être capable de
    réaliser en fonctionnement nominal avec un chargement de 500kgs.

2.  Calculez la masse équivalente des inerties des roues et du
    réducteur. Conclure quant à leurs impacts sur la dynamique du
    système.

> Retrouvez ce résultat sur le modèle *TestSlopeMotor.*

1.  On souhaite à présent modéliser un profil d’utilisation plus
    représentatif des conditions réelles d’utilisation du véhicule,
    complétez le modèle *TestProfile_Cycle.*

> Relevez la puissance maximale à fournir par le moteur pour réaliser
> cette mission et le couple RMS en fin de cycle.

<u>Remarque 3</u> : le bloc *SpeedTable* fournit l’évolution du profil
de vitesse au cours du temps et le bloc *SlopeTable* l’évolution de la
pente en fonction de la position du véhicule.

<u>Remarque 4</u> : le Root Mean Square Torque (couple RMS) est la
moyenne quadratique du couple au cours d’un cycle. Pour des
considérations thermiques, il faut que ce couple soit inférieur au
couple nominal du moteur.

1.  Établir le profil vitesse de rotation et couple moteur pour réaliser
    ce cycle.

<u>Remarque 5</u> : Sur Dymola, après simulation, vous pouvez changer la
variable des abscisses de votre graphe en utilisant l'outil *Indepent
Variable* de l’onglet *Plot Options*. Vous devez au préalable avoir
sélectionné dans l’arborescence *Variable Browser* la variable que vous
souhaitez utiliser comme référence.
